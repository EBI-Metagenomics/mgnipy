#  &#x1F575; Exploring MGnify Resources &#x1F5C2;

The MGnify API provides access to multiple types of resources (or endpoints) such as studies, samples, analyses, runs, and more... 

In this notebook we show you how to:

1. **Discover** what resources are available
2. **Inspect** what parameters each resource accepts
3. **Query** resources using two different approaches



## &#x270C; Two Ways to Query Resources 

MGnipy provides two main interfaces:
- **`MGnipy` client**: High-level interface with built-in helper functions for exploration
- **Resource proxies** (`mgnipy.V2.proxies`): Direct access to individual resource types

Let's start by exploring available resources using the `MGnipy` client.

---



### 1. `MGnipy` Client

```python
from mgnipy import MGnipy 

# init 
MG = MGnipy(
    # add a configuration 
)

# we can explore which resources are available
MG.list_resources()
```

For more detail we can describe the resource

```python
print("studies list endpoint:")
MG.describe_resources("studies")

print("\n----------\n")

print("analysis detail endpoint:")
MG.describe_resources("analysis")
```

To use a given endpoint you can access it as an attribute

```python
studies = MG.studies
filtered_studies = studies.filter(search="chicken")
filtered_studies.explain()
# or 
print("\n----------\n")
filtered_studies = MG.studies(search="chicken")
filtered_studies.explain()
```

again to help there are helper functions for each resource proxy such as `.list_supported_params()` `.describe_endpoint()`

```python
print(studies.list_supported_params())
# or 
print("\n----------\n")
studies.describe_endpoint()
```

### 2. Resource `mgnipy.V2.proxies`

Alternatively, many of the same functionalities are available from the resource proxies

```python
from mgnipy.V2.proxies import Studies
```

```python
chicken_studies = Studies(search="chicken")
chicken_studies.describe_endpoint()
print("\n----------\n")
chicken_studies.explain()
```

## List &#x1F4DA; vs. Detail &#x1F4D6; proxies

In the avail resource (`MGnipy.list_resources()`), the plural endpoints are MGnify API list endpoints, while the singular are endpoints that can be used to retrieve details about a single resource. 

For example we will look at list proxy `genomes` vs. detail proxy `genome`. Again we could use either interface: mgnipy.MGnipy or the mgnipy.V2.proxies. 

We will access via MGnipy

### MGnifyList proxy

```python
# init 
genomes = MG.genomes
# let's take a look at the number of records
genomes.dry_run()
```

that's a lot. Let's not get all. Instead we can get the first 3 pages. 

```python
# get pages 1 to 3
genomes.get(pages=[1,2,3])

# and we can take a look at the results 
genomes.to_df().head()
```

### MGnifyDetail proxy

Let's say we want the details about the first genome i.e., `GenomeDetail`

We can access this from our already initiated `genomes` search instance (e.g. traversing relationships - see next guide)

OR pass our known accession e.g. `MGYG000314684` to the `genome` proxy

```python
genome_accession = "MGYG000314684"

genome_dets1 = genomes[genome_accession]
print(genome_dets1)

# or, same as
genome_dets2 = MG.genome(genome_accession)
print(genome_dets2)

# or, same as
from mgnipy.V2.proxies import GenomeDetail
genome_dets3 = GenomeDetail(genome_accession)
print(genome_dets3)
```
