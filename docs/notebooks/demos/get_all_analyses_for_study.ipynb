## 🔎 Getting the Study's details
Here we import and configuire our MGnipy client

```python
from mgnipy import MGnipy
import asyncio # optional, for async requests

# init
MG = MGnipy(
    #cache_dir="cache" # default dir is platform-specific,set to None to disable caching
)

# access study resource
study = MG.study(study_accession)

# check out call before making it
study.dry_run()
```

let's go ahead and get the study's details:

```python
# get
study.get()

# can clear cache now optionally or later or never
study.clear_cache()

# stored to results, so can view as list, dfs etc
study.to_df()
```

```python
study.downloads
```

## 📒 `MGazine` for accessing the study's datasets

- study details have a `mgnipy.MGazine` which allow us to download and interact with study-level datasets outputed from MGnify. 

- We can use `mgnipy.MGazine` to download the datasets onto disk or read them into our notebook. 

- To access the study's mgazine use `.datasets` 

- the str representaiton of mgazine tells us the number of download files and their {alias: url}

```python
# getting mgazine
mgazine = study.datasets
# see info
print(mgazine)
```

### ⬇️ Downloading a copy of the files using `.download()` or `.download_all()`
- also with async support
- for downloading only need to pass a destination dir and the file alias (helper list `.aliases`) or url (helpers `.url_list`, `.url_dict`)

```python
# check out file aliases
aa = mgazine.aliases

aa
```

```python
# download one file by alias
mgazine.download(to_dir="downloads", alias=aa[1])
```

```python
# downloading all, async optional
await mgazine.adownload_all(to_dir="downloads")
```

### 📖 Alternatively, reading in the files using `stream`

- to continue working with the datasets before downloading
- support for csv/tsv (gzipped or not), txt, fasta, gff, biom, json, jsonl (see `mgnipy.V2.mixins.StreamMixin`)

```python
mgazine.stream(alias=aa[1])
```
