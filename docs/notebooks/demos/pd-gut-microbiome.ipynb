# Demo: Integrating multiple studies

In this notebook we will try to recreate the analyses: 

[ABaCo demo: Parkinson's disease gut microbiome](https://mona-abaco.readthedocs.io/en/latest/tutorial/demo-parkinson.html)

> _This notebook is used as a demo of ABaCo's capabilities on integrating multiple studies with multiple biomes. In this case, the Parkinson's dataset contains samples from the gut microbiome from 5 different studies._

but starting from mgnipy to get taxa tables and metadata

````python
from mgnipy import MGnipy

# configuring
mg = MGnipy(
    # configuration
    cache_dir="cache",
)
````

## Finding the studies

````python
# search for parkinson's disease gut microbiome studies
mg_pd = mg.studies(
    biome_lineage='root:Host-associated:Human:Digestive system:Large intestine:Fecal',
    search=["parkinson's"]
)

# preview how many requests
mg_pd.explain()
````

now making the one get request to see the initial list of studies matching our above filter criteria

````python
# getting the initial list of pd studies
mg_pd.get()

# view results as df, optional 
mg_pd.to_df(expand_nested_dicts=True)
````

## Getting some Metadata

that's nice but let's take a look at the studies details. We can use `.enrich_details()` which will basically `.get_detail()` for all studies. 

Below we use the async version `.aenrich_details()`

````python
await mg_pd.aenrich_details(
    limit=10, #optional, number of details to enrich in this call
    hide_progress=False # optional, whether to hide progress bar
)
````

to check it out we can use `.details_results` or `.details_df()`

````python
mg_pd.details_df(expand_nested_dicts=True)
````

- Study and Analysis details have download fields. 
- In a given `Study` / `AnalysisDetail` you can take a closer look via their `.downloads` list
- in `MGnifyList`s such as `Studies` and `Analyses` all their detail downloads are combined into a `.details_downloads` list

````python
# checking out 0th in download
mg_pd.details_downloads[0]
````

there's also a `.downloads_df()`

````python
# accessing downloads_df and checking out pipeline versions
mg_pd.downloads_df['pipeline_version'].value_counts()
````

These downloads lists are used in `MGazine`s to download, read in and curate mgnify datasets with metadata. 

The `MGazine` for this `Studies` list i.e., `mg_pd` can be accessed via `.datasets` attribute

## Retrieving the datasets

````python
# getting mgazine
mz = mg_pd.datasets

# we can print more info about the mgazine dataset
print(mz)
````

Further filtered mgazine datasets or curations can be accessed via:
- calling attributes corresponding to specific pipeline versions i.e., `.v3`
- using the short description as a subscript i.e., `["Complete GO annotation"]`

for example let's say we are interested in 'Taxonomic assignments SSU' from pipeline 4.1

````python
# checking out the v4.1 datasets
print(mz.v4_1)
````

````python
# accessing the taxonomic assignments SSU dataset in v4.1
tax_ssu_cur = mz.v4_1['Taxonomic assignments SSU']
# sanity check
print(tax_ssu_cur)
````

we can get out the curated taxonomic datasets as

- `to_pandas`()
- `to_polars()` as polars dataframe
    - `lazy_merged` for polars lazyframe
- `X()` for array of values only, to either polars or pandas
- `metadata()` for mgnify enriched runs/study/sample etc metadata, to either polars or pandas
- `taxonomic_metadata()` for taxonomic ranks data, to either polars or pandas
- `to_anndata()` **(not for DwC-ready)**
    - its data `X` from `X()`
    - `var` annotations from `metadata()`
    - `obs` annotations form `taxonomic_metadata()`

![Note] check out the getting started with anndata page for an intro

````python
# getting the annotated dataframe
ad = tax_ssu_cur.to_anndata()
print(ad)
# looking at the obs metadta
ad.obs.head()
````

- notice the warning:
    ```
    WARNING:root:No runs have been enriched yet. Returning empty metadata DataFrame. Please run `enrich_runs()` first to populate the metadata.
    ```

- and that there is no `var` in the anndataframe: 
    ```
    AnnData object with n_obs × n_vars = 2286 × 413
        obs: 'Superkingdom', 'Kingdom', 'Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species'
    ```

- we can enrich with metadata using `enrich_runs()` which will call a bunch of `RunDetails` 

- alternatively run level metadata can be passed on to `runs_results: list[dict] = ...` 

````python
tax_ssu_cur.enrich_runs(limit=None)
````

````python
ad_more = tax_ssu_cur.to_anndata()
print(ad_more)
ad_more.var.head()
````
